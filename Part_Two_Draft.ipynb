```python
#schaffer
def schaffer(X):
    x, y = X[:, 0], X[:, 1]
    return 0.5 + ((np.sin(x**2 - y**2))**2 - 0.5) / ((1 + 0.001 * (x**2 + y**2))**2)
    
#modified ackley
def modackley(X):
    x, y = X[:, 0], X[:, 1]
    return -20 * np.exp(-0.2 * np.sqrt(0.5 * (x**2 + y**2))) \
           - np.exp(0.5 * (np.cos(2 * np.pi * x) + np.cos(2 * np.pi * y))) + 20 + np.e
#easom
def easom(X):
    x, y = X[:, 0], X[:, 1]
    return -np.cos(x) * np.cos(y) * np.exp(-((x - np.pi)**2 + (y - np.pi)**2))

problems = [schaffer, modackley, easom]
```

```python
import numpy as np
import matplotlib.pyplot as plt

search_constraints = (-5, 5)
random_solutions = []

def plot_fitness():
    for f in problems:
        fitness = f(random_solutions) 
        plt.scatter(random_solutions[:,0],random_solutions[:,1],c = fitness)
        plt.colorbar ()
        plt.show()

for r in range(5000):
    x = np.random.uniform(*search_constraints)
    y = np.random.uniform(*search_constraints)
    random_solutions.append([x, y])

random_solutions = np.array(random_solutions)
plot_fitness()
```
